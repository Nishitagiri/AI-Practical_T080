```python
# A* Search Algorithm
# MVLU College -> Priyadarshini Park

from queue import PriorityQueue
import heapq
import matplotlib.pyplot as plt
import networkx as nx

#data structure population

graph = {
    'MVLU College': {'Bandra': 12, 'Sion': 14},
    'Bandra': {'Worli': 6, 'Byculla': 10},
    'Sion': {'Byculla': 13},
    'Worli': {'Byculla': 5, 'Haji Ali': 8},
    'Byculla': {'Haji Ali': 6},
    'Haji Ali': {'Priyadarshini Park': 5},
    'Priyadarshini Park': {}
}

heuristics = {
    'MVLU College': 24,
    'Bandra': 15,
    'Sion': 16,
    'Worli': 8,
    'Byculla': 7,
    'Haji Ali': 3,
    'Priyadarshini Park': 0
}

node_positions = {
    'MVLU College': (2,8),
    'Bandra': (1,6),
    'Sion': (3,6),
    'Worli': (1,4),
    'Byculla': (3,4),
    'Haji Ali': (2,2),
    'Priyadarshini Park': (2,0)
}

start='MVLU College'
goal='Priyadarshini Park'


print("========== A* SEARCH EXPANSION ==========\n")

pq=PriorityQueue()
pq.put((heuristics[start],0,[start]))

visited=set()

while not pq.empty():

    f,g,path=pq.get()

    node=path[-1]

    if node==goal:

        print("\nBest Path:"," -> ".join(path))
        print("Total Cost =",g)
        break

    if node in visited:
        continue

    visited.add(node)

    for neighbour,cost in graph[node].items():

        new_g=g+cost
        new_f=new_g+heuristics[neighbour]

        print(path+[neighbour],
              "g =",new_g,
              "h =",heuristics[neighbour],
              "f =",new_f)

        pq.put((new_f,new_g,path+[neighbour]))

#A* search function

def a_star_search(graph,heuristics,start,goal):

    priority_queue=[(heuristics[start],start,[start],0)]

    visited=set()

    while priority_queue:

        f_score,current,path,g_score=heapq.heappop(priority_queue)

        if current in visited:
            continue

        visited.add(current)

        if current==goal:
            return path,g_score

        for neighbour,distance in graph[current].items():

            if neighbour not in visited:

                new_g=g_score+distance
                new_f=new_g+heuristics[neighbour]

                heapq.heappush(
                    priority_queue,
                    (new_f,neighbour,path+[neighbour],new_g)
                )

    return None,float('inf')

#run A*

optimal_path,total_distance=a_star_search(
    graph,
    heuristics,
    start,
    goal
)

print("\n========== OPTIMAL PATH ==========\n")
print("Optimal Path:")
print(" -> ".join(optimal_path))

print("\nTotal Road Distance:",total_distance,"km")

#graph

G=nx.DiGraph()

for node,neighbours in graph.items():

    for neighbour,weight in neighbours.items():

        G.add_edge(node,neighbour,weight=weight)

plt.figure(figsize=(10,8))

path_edges=list(zip(optimal_path,optimal_path[1:]))

normal_edges=[
    edge for edge in G.edges()
    if edge not in path_edges
]

nx.draw_networkx_nodes(
    G,
    node_positions,
    node_size=2500,
    node_color='lightblue'
)

nx.draw_networkx_edges(
    G,
    node_positions,
    edgelist=normal_edges,
    edge_color='gray',
    arrows=True
)

nx.draw_networkx_edges(
    G,
    node_positions,
    edgelist=path_edges,
    edge_color='darkorange',
    width=3,
    arrows=True
)

labels={
    node:f"{node}\nh(n)={heuristics[node]}"
    for node in G.nodes()
}

nx.draw_networkx_labels(
    G,
    node_positions,
    labels,
    font_size=9,
    font_weight='bold'
)

edge_labels=nx.get_edge_attributes(G,'weight')

edge_labels={
    k:f"{v} km"
    for k,v in edge_labels.items()
}

nx.draw_networkx_edge_labels(
    G,
    node_positions,
    edge_labels=edge_labels,
    font_color='red'
)

plt.title(
    "A* Search : MVLU College to Priyadarshini Park '080\n"
    "(Orange = Optimal Path)",
    fontsize=12
)

plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# Recursive Best-First Search (RBFS)
# MVLU College -> Priyadarshini Park

import matplotlib.pyplot as plt

#network definition

graph = {
    'MVLU College': {'Bandra': 12, 'Sion': 14},
    'Bandra': {'Worli': 6, 'Byculla': 10},
    'Sion': {'Byculla': 13},
    'Worli': {'Byculla': 5, 'Haji Ali': 8},
    'Byculla': {'Haji Ali': 6},
    'Haji Ali': {'Priyadarshini Park': 5},
    'Priyadarshini Park': {}
}

heuristics = {
    'MVLU College': 24,
    'Bandra': 15,
    'Sion': 16,
    'Worli': 8,
    'Byculla': 7,
    'Haji Ali': 3,
    'Priyadarshini Park': 0
}

coords = {
    'MVLU College': (2,6),
    'Bandra': (1,4.5),
    'Sion': (3,4.5),
    'Worli': (1,3),
    'Byculla': (3,3),
    'Haji Ali': (2,1.5),
    'Priyadarshini Park': (2,0)
}

#RBFS

def rbfs(node, goal, g, f_limit, path):

    print("Current Path:", path,
          "g =", g,
          "h =", heuristics[node],
          "f =", g + heuristics[node])

    if node == goal:
        return True, path, g, g

    neighbours = graph[node]

    if not neighbours:
        return False, [], 0, float('inf')

    successors = []

    for neighbour, distance in neighbours.items():

        if neighbour not in path:

            next_g = g + distance
            next_f = max(next_g + heuristics[neighbour],
                         g + heuristics[node])

            successors.append([next_f, neighbour, next_g])

    if not successors:
        return False, [], 0, float('inf')

    while True:

        successors.sort(key=lambda x: x[0])

        best = successors[0]

        if best[0] > f_limit:
            return False, [], 0, best[0]

        alternative = successors[1][0] if len(successors) > 1 else float('inf')

        success, result_path, total_g, returned_f = rbfs(
            best[1],
            goal,
            best[2],
            min(f_limit, alternative),
            path + [best[1]]
        )

        best[0] = returned_f

        if success:
            return True, result_path, total_g, returned_f


#run RBFS

print("========== RECURSIVE BEST-FIRST SEARCH ==========\n")

success, path, total_dist, _ = rbfs(
    'MVLU College',
    'Priyadarshini Park',
    g=0,
    f_limit=float('inf'),
    path=['MVLU College']
)

print("\n========== OPTIMAL PATH ==========\n")
print("Optimal RBFS Path:")
print(" -> ".join(path))
print("Total Distance:", total_dist, "km")

#graph plotting

plt.figure(figsize=(8,7))

for node, neighbours in graph.items():

    x1, y1 = coords[node]

    for neighbour, dist in neighbours.items():

        x2, y2 = coords[neighbour]

        is_path = (
            node in path and
            neighbour in path and
            path.index(neighbour) == path.index(node) + 1
        )

        color = "#2ecc71" if is_path else "#bdc3c7"
        width = 3 if is_path else 1.5

        plt.plot(
            [x1, x2],
            [y1, y2],
            color=color,
            linewidth=width
        )

        plt.text(
            (x1+x2)/2,
            (y1+y2)/2,
            str(dist) + " km",
            color="red",
            fontsize=9
        )

for node, (x, y) in coords.items():

    color = "#f1c40f" if node in path else "#3498db"

    plt.scatter(
        x,
        y,
        s=700,
        color=color,
        zorder=2
    )

    plt.text(
        x,
        y,
        node + "\nh=" + str(heuristics[node]),
        ha='center',
        va='center',
        color='black',
        fontsize=8,
        fontweight='bold'
    )

plt.title(
    "Recursive Best-First Search\n"
    "MVLU College → Priyadarshini Park '080",
    fontsize=12,
    fontweight='bold'
)

plt.axis("off")
plt.show()
```
